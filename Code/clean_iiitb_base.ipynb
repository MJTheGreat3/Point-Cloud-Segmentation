```python
input_file = "IIIT_Base - Target_Readable.ply"
output_rgb_class = "IIITB_Base_RGB.ply"
output_class = "IIITB_Base.ply"
```

```python
with open(input_file, 'r') as fin, \
     open(output_rgb_class, 'w') as fout_rgb_class, \
     open(output_class, 'w') as fout_class:

    # Skip the header
    for line in fin:
        if line.strip() == "end_header":
            break

    for line in fin:
        parts = line.strip().split()
        if len(parts) < 9:
            continue  # skip malformed lines

        # Extract relevant columns
        x, y, z = parts[0:3]
        r, g, b = parts[3:6]
        classification = parts[8]  # 9th column (0-based index 8)

        # Write x, y, z, r, g, b, classification
        fout_rgb_class.write(f"{x} {y} {z} {r} {g} {b} {classification}\n")

        # Write x, y, z, classification
        fout_class.write(f"{x} {y} {z} {classification}\n")

print("✅ Done! Files written successfully.")
```

```python
# Function to extract unique classification values from a PLY file
def extract_unique_classifications(ply_file):
    classifications = set()
    with open(ply_file, 'r') as file:
        lines = file.readlines()
        
        # Start processing after the header
        start_reading = False
        for line in lines:
            # Check if we've reached the vertex data section
            if start_reading:
                # Split the line and extract the scalar classification value
                parts = line.split()
                if len(parts) == 4:  # Assuming the format is: x, y, z, classification
                    classification_value = float(parts[3])
                    classifications.add(classification_value)
            # Mark where the vertex data starts
            if line.strip() == 'end_header':
                start_reading = True
    
    return classifications

# Example usage
ply_file_path = "IIITB_Base.ply"
unique_classifications = extract_unique_classifications(ply_file_path)

# Print the unique classification values
print("Unique Classification Values:")
for classification in unique_classifications:
    print(classification)
```
