```python
import os

# Directory containing the files ('.' means current directory)
directory = '.'

# Prefix to add
prefix = 'Campus_BlockA_'

# Loop through all files in the directory
for filename in os.listdir(directory):
    # Check if the file matches the pattern "group_<number>.ply"
    if filename.startswith('group_') and filename.endswith('.ply'):
        new_name = prefix + filename
        old_path = os.path.join(directory, filename)
        new_path = os.path.join(directory, new_name)

        # Rename the file
        os.rename(old_path, new_path)
        print(f"Renamed: {filename} → {new_name}")

print("✅ All matching files have been renamed.")
```
